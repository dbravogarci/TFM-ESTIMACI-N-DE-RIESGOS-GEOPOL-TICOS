```python
import os
import random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score, mean_absolute_percentage_error
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.optimizers import Adam
import tensorflow as tf
from tensorflow.keras.backend import clear_session

# Cargar datos
df = pd.read_csv("df_reduced_var_processed.csv", parse_dates=['week'])
df = df.sort_values("week").reset_index(drop=True)
df.set_index("week", inplace=True)

# Enmascarar NaNs
cols_with_nan = [
    'conflict_count', 'OPEP_involved', 'OPEP_conflict_count',
    'USA_involved', 'Iran_involved', 'Iran_conflict_count', 'shock_event'
]
for col in cols_with_nan:
    df[f'{col}_mask'] = df[col].isna().astype(int)
df[cols_with_nan] = df[cols_with_nan].fillna(0)

# Funci√≥n principal de entrenamiento
def run_lstm_prediction(df, target_var, exclude_targets, seed=42):
    os.environ['PYTHONHASHSEED'] = str(seed)
    os.environ['TF_DETERMINISTIC_OPS'] = '1'
    random.seed(seed)
    np.random.seed(seed)
    tf.random.set_seed(seed)
    clear_session()

    input_df = df.drop(columns=exclude_targets)
    scaler = MinMaxScaler()
    scaled_data = scaler.fit_transform(input_df)
    scaled_df = pd.DataFrame(scaled_data, columns=input_df.columns, index=df.index)

    def create_sequences(data, target_col, window_size=10):
        X, y = [], []
        for i in range(len(data) - window_size):
            X.append(data.iloc[i:i+window_size].values)
            y.append(data.iloc[i+window_size][target_col])
        return np.array(X), np.array(y)

    X, y = create_sequences(scaled_df, target_col=target_var)
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)

    model = Sequential()
    model.add(LSTM(8, activation='tanh', return_sequences=False, input_shape=(X.shape[1], X.shape[2])))
    model.add(Dropout(0.1458))
    model.add(Dense(1))
    model.compile(optimizer=Adam(learning_rate=0.00336), loss='mse')

    history = model.fit(X_train, y_train, epochs=50, batch_size=32, validation_data=(X_test, y_test), verbose=0)

    y_train_pred = model.predict(X_train).flatten()
    y_test_pred = model.predict(X_test).flatten()

    def regression_metrics(y_true, y_pred):
        mae = mean_absolute_error(y_true, y_pred)
        mse = mean_squared_error(y_true, y_pred)
        rmse = np.sqrt(mse)
        r2 = r2_score(y_true, y_pred)
        mape = mean_absolute_percentage_error(y_true, y_pred) * 100

        # SAFE-MAPE
        denominator = np.abs(y_true) + np.abs(y_pred)
        safe_mape = np.mean(np.where(denominator == 0, 0, 2 * np.abs(y_pred - y_true) / denominator)) * 100

        return mae, mse, rmse, r2, mape, safe_mape

    train_metrics = regression_metrics(y_train, y_train_pred)
    test_metrics = regression_metrics(y_test, y_test_pred)

    metrics_df = pd.DataFrame({
        "Conjunto": ["Entrenamiento", "Validaci√≥n"],
        "MAE": [train_metrics[0], test_metrics[0]],
        "MSE": [train_metrics[1], test_metrics[1]],
        "RMSE": [train_metrics[2], test_metrics[2]],
        "R¬≤": [train_metrics[3], test_metrics[3]],
        "MAPE (%)": [train_metrics[4], test_metrics[4]],
        "SAFE-MAPE (%)": [train_metrics[5], test_metrics[5]]
    })


    # Mostrar curvas de p√©rdida
    plt.figure(figsize=(6, 4))
    plt.plot(history.history['loss'], label='Training Loss')
    plt.plot(history.history['val_loss'], label='Validation Loss')
    plt.title(f'Loss Curve ‚Äì {target_var}')
    plt.xlabel('Epoch')
    plt.ylabel('MSE')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

    return model, history, metrics_df

# Excluir columnas seg√∫n objetivo
gold_exclude = ['close_oil_norm', 'close_gas_norm']
oil_exclude = ['close_gold_norm', 'close_gas_norm']
gas_exclude = ['close_gold_norm', 'close_oil_norm']

# Entrenar modelos
model_gold, _, metrics_gold = run_lstm_prediction(df, 'close_gold_norm', gold_exclude)
model_oil, _, metrics_oil = run_lstm_prediction(df, 'close_oil_norm', oil_exclude)
model_gas, _, metrics_gas = run_lstm_prediction(df, 'close_gas_norm', gas_exclude)

print("üî∂ Oro:\n", metrics_gold)
print("üî∑ Petr√≥leo:\n", metrics_oil)
print("üü© Gas:\n", metrics_gas)

# Fechas de eventos
eventos = {
    "2011-03-15": "Guerra en Siria",
    "2013-07-03": "Golpe en Egipto",
    "2014-06-29": "ISIS Califato",
    "2015-03-26": "Intervenci√≥n en Yemen",
    "2016-07-15": "Golpe en Turqu√≠a",
    "2017-06-05": "Ruptura con Qatar",
    "2019-06-20": "Ir√°n derriba dron",
    "2020-01-03": "Soleimani asesinado",
    "2021-03-23": "Bloqueo del Canal de Suez",
    "2022-02-24": "Inicio guerra en Ucrania",
    "2023-10-07": "Ataque de Ham√°s",
    "2023-11-19": "Crisis Mar Rojo (ataques Houthi)"
}


# Visualizar predicciones globales para cada materia prima
def plot_global_predictions(model, df, target_col, exclude_cols, eventos, window=10):
    input_df = df.drop(columns=exclude_cols)
    scaler = MinMaxScaler()
    scaled_data = scaler.fit_transform(input_df)
    scaled_df = pd.DataFrame(scaled_data, columns=input_df.columns, index=df.index)

    def create_sequences(data, target_col, window_size=10):
        X, y, indices = [], [], []
        for i in range(len(data) - window_size):
            X.append(data.iloc[i:i+window_size].values)
            y.append(data.iloc[i+window_size][target_col])
            indices.append(data.index[i+window_size])
        return np.array(X), np.array(y), indices

    X, y, idx = create_sequences(scaled_df, target_col=target_col)
    y_pred = model.predict(X).flatten()
    idx = pd.to_datetime(idx)
    series_pred = pd.Series(y_pred, index=idx)
    series_true = pd.Series(y, index=idx)

    # Gr√°fica
    plt.figure(figsize=(12, 5))
    plt.plot(series_true, label='Real', color='blue', alpha=0.7)
    plt.plot(series_pred, label='Predicci√≥n', color='orange', alpha=0.7)

    for fecha, nombre in eventos.items():
        event_dt = pd.to_datetime(fecha)
        if series_true.index.min() <= event_dt <= series_true.index.max():
            plt.axvline(event_dt, color='red', linestyle='--', alpha=0.5)
            plt.text(event_dt, plt.ylim()[1]*0.98, nombre, rotation=90, fontsize=7, color='red', ha='right', va='top')


    plt.title(f"Predicci√≥n de {target_col}")
    plt.xlabel("Fecha")
    plt.ylabel("Valor Normalizado")
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

# Mostrar predicciones globales para cada materia prima
plot_global_predictions(model_gold, df, 'close_gold_norm', gold_exclude, eventos)
plot_global_predictions(model_oil, df, 'close_oil_norm', oil_exclude, eventos)
plot_global_predictions(model_gas, df, 'close_gas_norm', gas_exclude, eventos)
```
